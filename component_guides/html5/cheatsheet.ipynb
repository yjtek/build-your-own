# HTML

- ```<head>```
    - ```<meta name='' content=''>``` 
        - Put in page metadata 
    - ```<title>```
        - Title displayed in tab
    - ```<link rel="icon" href="html5.png" type="image/x-icon">``` OR ```<link rel="stylesheet" href="main.css" type="text/css">```
        - Link to tab icon
        - Link to css page

- ```<body>```
    - ```<h1/2/3/4/5>```
        - Title text

    - ```<hr>```
        - Horizontal rule
    
    - ```<p>```
        - paragraph

    - ```
        <ol>
            <li>
        ```
        - Ordered list with list item

    - ```
        <ul>
            <li>
        ```
        - Unordered list with list item
    
    - ```
        <dl>
            <dt>
                <dd>
        ```
        - Description list, description term, and descriptions

    - ```<a href="about.html">Dave Gray</a>```
        - Add a hyperlink to a given text

    - ```
        <nav>
            <ul>
                <li>
                    <a href="#html">Learning HTML</a>
                </li>
                <li>
                    <a href="#vacation">Planning a Vacation</a>
                </li>
            </ul>
        </nav>
        ```
        - Adding hyperlinks as content page navigation section

    - ```
        <figure>
            <img src="img/vacation.jpg" alt="Cancun Vacation" title="It's 5 o'Clock Somewhere!" width="400"
                height="267" loading="lazy">
            <figcaption>
                A Caribbean Vacation Image
            </figcaption>
        </figure>
        ```
        - `alt` loads text if picture not found, `title` is what you get if you mouseover the picture, `figcaption` is text that appears under pic, others are self explanatroy

     - ```<header>``` / ```<main>``` / ```<footer>```
        - Semantic elements; convenient way to group stuff that needs to appear at the top / middle / bottom 
    
    - ```<article id="html">``` / ```<section>``` / ```<nav>``` / ```<aside>```
        - Used to name a chunk of content. You can reference this in a hyperlink by name using ```<a href="#html">```
        - article > section > nav > aside in terms of sectioning size. i.e. section cannot contain article, but article can contain section

    - ```<details>```
        - Used to create a foldable disclosure section 

    - ```table, tr, th, td, caption```
        - Table components
        - Caption is just the table title
    
    - ```<form action="https://httpbin.org/get" method="get">```
        - do a GET request to specific URL

    - ```<fieldset>```
        - group related elements within a form and create a visual distinction or boundary around them
        
    - ```<button type="submit" formaction="https://httpbin.org/post" formmethod="post">Post</button>```

# CSS

- Segment before curly braces is the component type, segment within curly braces are parameters


- Parameters
    - font-size
    - background-color (page background color)
    - color
        - text colour
    - border
    - font-family
    - border-collapse
        - If you have table within tables, don't show inner borders and display as seamless table
    - padding
